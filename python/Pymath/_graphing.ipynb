```python

import numpy as np
import sympy as sp
import matplotlib.pyplot as plt
from IPython.display import display, Math
```

```python
fig, ax = plt.subplots()

L = 4
x = np.linspace(0, L)
ncolors = len(plt.rcParams['axes.prop_cycle'])
shift = np.linspace(0, L, ncolors, endpoint=False)
for s in shift:
    ax.plot(x, np.sin(x + s), '')

plt.show()
```

# EXERCISE 1

```python
for x in range(-10,11):
    y = x**2 - 3*x
    plt.plot(x,y,'o')

plt.grid()
plt.show()
```

```python
# MIKE'S SOLUTION
x = sp.symbols('x')
y = x**2 - 3*x

for i in range(-11,12):
    plt.plot(i,y.subs(x,i),'o')

plt.xlabel('x')
plt.ylabel('f(x) = %s' %sp.latex(y))

plt.grid()
plt.show()  
```


## Slope-intercept Form

```python
#y = mx+b

x = [-5,5]
m = .7
b = -2

y = [0,0]

# for i in range(len(x)):
#     y[i] = m*x[i] + b

y = m*np.array(x)+b
    
print(x,y)

plt.plot(x,y,label='y=%sx+%s'%(m,b))

plt.grid()
plt.xlim(x)
plt.ylim(x)

axis = plt.gca()

plt.plot(axis.get_xlim(),[0,0],'k--')
plt.plot([0,0],axis.get_ylim(),'k--')

plt.legend()

plt.plot(x,y)

plt.show()
```

```python

x = [-5,5,6,3]
m = [0.7, -1.25,4,1]
b = [-2,0.75,5,2]
y = [0,0,0]

for i in range(len(x)):
    y = m[i]*np.array(x)+b[i]
    plt.plot(x,y,label = "y= %sx+%s"%(m[i],b[i]))

plt.axis('square')
plt.xlim(-5,6)
plt.ylim(-5,6)

plt.plot([0,0],[-5,6],'k--')
plt.plot([-5,6],[0,0],'k--')

plt.grid()
plt.legend()

plt.show()

```

```python
x = [-5,5]
m = [.7,-1.25]
b = [-2,.75]

for i in range(len(x)):
    y = m[i]*np.array(x)+b[i]
    plt.plot(x,y)

plt.axis('square')
plt.xlim(-5,5)
plt.ylim(-5,5)
     
plt.axhline(color = 'k')
plt.axvline(color = 'k')

plt.grid()
plt.legend()

plt.show()
```
